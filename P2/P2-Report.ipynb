```python
# Reading the first 5 rows of data to uderstand the data provided using pandas

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns


df = pd.read_csv('titanic_data.csv')

df.head(20)
```

From the loaded data we can see that 'Age' is not arranged in order... We now will try sorting the data 

```python
#Sorting Values by Age

sorted_frame = df.sort_values(['Age'])

# Printing first five rows of sorted values 

sorted_frame.head(20)
```


From the loaded data we can see there are some null values in the 'Age' variable in the data... We now will try cleaning the data by deleting the null values from the data 


```python
# Removing null values from Age Column and Saving the frame as New_Age

New_Age = df[pd.notnull(df['Age'])]

New_Age.head(10)
```

We will now sort the data without nulls 

```python
#Sorting After deleting null values

New_Age.sort_values('Age').head(20)
```

We will now delete the columns which wee will not be using for our analysis based on the questions 

```python
#Cleaning data which will not be used

Clean_Data = New_Age.drop(['PassengerId','SibSp','Cabin','Parch','Name','Ticket'], axis=1)
```

```python
Clean_Data.head()
```

Grouping the Data Acoording to the Passenger Class and the Survived will gives us some overview of the data which might give us insight into the relation between survived and passenger class. but before we need to find out the total number of survived passengers to compare the passenger class

```python
#Grouping the data in order to find the number of survived


Clean_Data.groupby(['Survived']).count()
```

Grouping the data Acoording to Pclass and Survived to find the relation betwwen them 

```python
Clean_Data.groupby(['Survived','Pclass']).count()
```

After arranging the data we can clearly see that there is some relation between the Passenger Class and the Survival rate . Now we arrange the data according to sex will provide us details of the relation between the sex and the number of survived 

```python
#Grouping the Clean_data in order to find the relation between number of survived and Age,Sex

Clean_Data.groupby(['Survived','Sex','Pclass']).count()
```

By Grouping the data we can clearly see that the rate of female survivors is more than that of male survivors ...One Reason for this might be that the number of female passengers are more than that of male on the titanic... but according to original data we can observe that male count is greater than female and thus we can forget the assumption of females are more than males on titanic. But by grouping the data we can clearly see that there is some relation between Passenger Class and the number of survivors



We now will try to find the co-relation between different variables in order to find the effects of each other on different variables 

```python
# Standerdizing Age and other variables 

# Caluculating Co-Relation between variables i.e Pearson's r

def correlation(x,y) :
    std_x = (x-x.mean()) / x.std(ddof=0)
    std_y = (y-y.mean()) / y.std(ddof=0)
    
    return (std_x * std_y ).mean()
```

```python
correlation(Clean_Data['Age'], Clean_Data['Pclass'])
```

```python
correlation(Clean_Data['Age'], Clean_Data['Survived'])
```

```python
correlation(Clean_Data['Pclass'], Clean_Data['Survived'])
```

```python
correlation(Clean_Data['Survived'], Clean_Data['Fare'])
```

```python
#Drawing a scatter plot to better understand the relation between variables 

# define colors list, to be used to plot survived either red (=0) or green (=1)


%matplotlib inline

from pandas.tools.plotting import scatter_matrix

colors=['red','green']

# make a scatter plot

scatter_matrix(Clean_Data,figsize=[20,20],marker='x',c=Clean_Data.Survived.apply(lambda x:colors[x]))


Clean_Data.info()

```

```python
# Probability of survival by gender

Prob_Gen = Clean_Data.groupby('Sex')['Survived'].mean()
Prob_Gen
```

```python
# Probabilty of survival by gender chart

Prob_Gen.plot(kind='bar', title='Probability Of Survival by Gender', color=list('rgbkym'))
```

```python
# probability of survival by Passenger Class

Prob_Class = Clean_Data.groupby('Pclass')['Survived'].mean()
Prob_Class
```

```python
# Probabilty of survival by Passenger Class chart

Prob_Class.plot(kind='bar', stacked='True',  title='Probability Of Survival by Passenger Class', color=list('rgbkym'))
```

```python
#Probability of survival by both gender and class

Prob_gen_Class = Clean_Data.groupby(['Pclass','Sex'])['Survived'].mean()
Prob_gen_Class
```

```python
Prob_gen_Class.plot(kind='bar', title='Probability Of Survival by Gender and Passenger Class', color=list('rgbkym'))
```
